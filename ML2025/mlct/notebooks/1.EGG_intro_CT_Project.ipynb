# Intro de GD ITESO

$$
f(x) = x^2 + 1
$$
$$
f'(x)=2x
$$

```python

print("Lanzando GD test...")
initial_x = 5
alpha = 0.1
n_iter = 20

step_stop_threshold = 0.0000001

x = initial_x

iter_arr = [] # Iter array
y = [] # FunEvalArr

for i in range(n_iter):
    # Calculate gradient for actual x
    g = 2 * x
    # Update x with gradient
    x = x - alpha * g
    # Save the current value of f(x)
    y.append(x**2 + 1)
    iter_arr.append(i + 1)
    print(f"i: {i + 1}, x = {x}, y/f(x) = {x**2 + 1}")

    # if len(y) > 2:

    #     if abs(y[-1] - y[-2]) < step_stop_threshold:
    #         # print(f'y[-1] = {y[-1]}')
    #         # print(f'y[-2] = {y[-2]}')
    #         # print(f'y[-1] - y[-2] = {y[-1] - y[-2]}')
    #         break
```

```python
import matplotlib.pyplot as plt
from numpy import linspace

plt.subplot(1, 2, 1)
plt.plot(iter_arr, y)
plt.xlabel("Iteration/Epoch")
plt.ylabel("Loss/y/f(x)/J/C")

plt.subplot(1, 2, 2)
X = linspace(-5, 5, 100)
Y = X**2 + 1
plt.plot(X, Y, x, y[-1], "ro")
plt.xlabel("X")
plt.ylabel("Y")
```
