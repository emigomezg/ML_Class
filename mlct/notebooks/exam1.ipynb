1. Non-linear data can be approximated using multiple linear models.

### R: True

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

CSV_PATH = r"C:\\Users\\emi_g\\Downloads\\exam1_data.csv"
df = pd.read_csv(CSV_PATH, na_values='?')

X = df['x']
y = df['y']

plt.scatter(X, y, label='Data', color='gray')

# 0-20
mask1 = (X >= 0) & (X <= 20)
X1 = X[mask1].values.reshape(-1, 1)
y1 = y[mask1]
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.2, random_state=42)
model1 = LinearRegression()
model1.fit(X1_train, y1_train)
y1_pred = model1.predict(X1)
plt.plot(X[mask1], y1_pred, color='red', label='Fit 0-20')
print("R2 for 0-20 (test):", model1.score(X1_test, y1_test))

# 21-40
mask2 = (X >= 21) & (X <= 40)
X2 = X[mask2].values.reshape(-1, 1)
y2 = y[mask2]
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.2, random_state=42)
model2 = LinearRegression()
model2.fit(X2_train, y2_train)
y2_pred = model2.predict(X2)
plt.plot(X[mask2], y2_pred, color='green', label='Fit 21-40')
print("R2 for 21-40 (test):", model2.score(X2_test, y2_test))

# 41-60
mask3 = (X >= 41) & (X <= 60)
X3 = X[mask3].values.reshape(-1, 1)
y3 = y[mask3]
X3_train, X3_test, y3_train, y3_test = train_test_split(X3, y3, test_size=0.2, random_state=42)
model3 = LinearRegression()
model3.fit(X3_train, y3_train)
y3_pred = model3.predict(X3)
plt.plot(X[mask3], y3_pred, color='blue', label='Fit 41-60')
print("R2 for 41-60 (test):", model3.score(X3_test, y3_test))

plt.legend()
plt.show()
```
