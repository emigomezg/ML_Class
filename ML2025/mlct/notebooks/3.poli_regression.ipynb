$$
y = \frac{{1}}{{2}} x^2 + X + 2
$$

```python
import numpy as np

m = 100
X = 6 * np.random.rand(m, 1) - 3
y = 0.5 * X**2 + X + 2 + np.random.randn(m, 1)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y)
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=2)
X_ext = poly_features.fit_transform(X)

#SGDRegressor
from sklearn.linear_model import SGDRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

reg = make_pipeline(StandardScaler(),SGDRegressor(max_iter=1000, tol=1e-3))
reg.fit(X_ext, y.ravel())

X_new = np.linspace(-3, 3, 100).reshape(100, 1)
X_new_ext = poly_features.transform(X_new)
y_pred = reg.predict(X_new_ext)

# Plotting
plt.scatter(X, y, alpha=0.5, label='Real data')
plt.plot(X_new, y_pred, color='r', label='Prediction')

plt.xlabel('X')
plt.ylabel('y')
plt.title('Polynomial Regression with SGD')
plt.legend()
plt.show()
```

```python

from sklearn.linear_model import LinearRegression

#LinearRegression

reg2 = make_pipeline(StandardScaler(),LinearRegression())
reg2.fit(X_ext, y.ravel())

X_new2 = np.linspace(X.min(), X.max(), 100).reshape(100, 1)
X_new_ext2 = poly_features.transform(X_new2)
y_pred2 = reg2.predict(X_new_ext2)

# Plotting
plt.scatter(X, y, alpha=0.5, label='Real data')
plt.plot(X_new2, y_pred2, color='r', label='Prediction')

plt.xlabel('X')
plt.ylabel('y')
plt.title('Polynomial Regression with LinearRegression')
plt.legend()
plt.show()
```
